# Data Visualization

* Data visualization packages
    * matplotlib.pyplot
    * seaborn
* Plots
    * line plot: `plt.plot(x, y)`
    * scatter plot: `plt.scatter(x, y)`
    * bar plot: `plt.bar(categories, values)`
    * histogram: `plt.hist(data)`
    * error bars: `plt.errorbar(x, y, yerr=errorBar)`
    * violin plots: `sns.violinplot(df)`
* Formating
    * axis labels: `plt.xlable('x')` and `plt.ylabel('y')`
    * title: `plt.title('my plot')
    * subplot:
      ```python
      fig, axs = plt.subplots(2, 2, figsize=(6, 8))
      ```

## Import packages

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pandas as pd

sns.set_context('notebook')
sns.set_style('darkgrid')
````

## Generate some data for plotting

````python
# Generate 50 datapoints that are normally distributed
x = np.random.randn(50)
print(x)
````

## Scatter plot

````python
# Plot data in scatter plot
plt.plot(x,'.')
plt.title('A Random Scatter Plot X')
plt.xlabel('X')
plt.ylabel('Y')
````

## Histogram

````python
# Plot the histogram of the data
plt.hist(x,5)
plt.title('A Random Histogram X')
plt.xlabel('X')
plt.ylabel('Y')
````

````python
# Generate 1000 random data points
Y = np.random.randn(1000)
````

````python
# Plot the hitogram with 25 bins
plt.hist(Y,25)
plt.title('A Random Scatter Plot Y')
plt.xlabel('X')
plt.ylabel('Y')
````

## Boxplot

````python
# Make a boxplot of the data
plt.boxplot(x, vert=False, sym='*')
plt.title('Random Horizontal Boxplot X')
plt.xlabel('X')
plt.ylabel('Y')
````

## Errorbars

````python
# Generate data for x, y and errorBar
x = np.arange(5)
y = x**2
errorBar = x/2
````

````python
# Plot x, y with errorbar
plt.errorbar(x, y, yerr = errorBar, fmt = 'o')
````

## 2D Plot

````python
# Generate data
x = np.random.randn(200)
y = 10 + 0.5*x + np.random.randn(len(x))
````

````python
# Scatter plot
plt.scatter(x, y)
````

### Violinplot (seaborn)

````python
# Generate the data
nd = stats.norm
data = nd.rvs(size=(100))
nd2 = stats.norm(loc=3, scale=1.5)
data2 = nd2.rvs(size=(100))
````

````python
# Make a dataFrame
    # column names: Group1 and Group2
    # column data: data and data2
df = pd.DataFrame({'Group1':data, 'Group2': data2})
# Generate a violinplot for the dataFrame
sns.violinplot(df)
````
